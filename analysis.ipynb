```python
import numpy as np
from scipy.stats import pearsonr


# Calculate quadratic and cubic correlations
def quadratic_correlation(x, y):
    x_squared = x ** 2
    return pearsonr(x_squared, y)[0]

def cubic_correlation(x, y):
    x_cubed = x ** 3
    return pearsonr(x_cubed, y)[0]

quad_corr_in_deez_apple = quadratic_correlation(data['in_deezer_playlists'], data['in_apple_playlists'])
cubic_corr_in_deez_apple = cubic_correlation(data['in_deezer_playlists'], data['in_apple_playlists'])
quad_corr_in_apple_deez = quadratic_correlation(data['in_apple_playlists'], data['in_deezer_playlists'])
cubic_corr_in_apple_deez = cubic_correlation(data['in_apple_playlists'], data['in_deezer_playlists'])

quad_corr_in_spotify_apple = quadratic_correlation(data['in_spotify_playlists'], data['in_apple_playlists'])
cubic_corr_in_spotify_apple = cubic_correlation(data['in_spotify_playlists'], data['in_apple_playlists'])
quad_corr_in_apple_spotify = quadratic_correlation(data['in_apple_playlists'], data['in_spotify_playlists'])
cubic_corr_in_apple_spotify = cubic_correlation(data['in_apple_playlists'], data['in_spotify_playlists'])

quad_corr_in_spotify_deez = quadratic_correlation(data['in_spotify_playlists'], data['in_deezer_playlists'])
cubic_corr_in_spotify_deez = cubic_correlation(data['in_spotify_playlists'], data['in_deezer_playlists'])
quad_corr_in_deez_spotify = quadratic_correlation(data['in_deezer_playlists'], data['in_spotify_playlists'])
cubic_corr_in_deez_spotify = cubic_correlation(data['in_deezer_playlists'], data['in_spotify_playlists'])

company_name_list = {"apple": "apple", "deez": "deez", "spotify": "spotify"}
for name in company_name_list:
    pass
print("Quadratic Correlation between deez and apple:", quad_corr_in_deez_apple)
print("Cubic Correlation between deez and apple:", cubic_corr_in_deez_apple)
print("Quadratic Correlation between apple and deez:", quad_corr_in_apple_deez)
print("Cubic Correlation between apple and deez:", cubic_corr_in_apple_deez)
print("Quadratic Correlation between spotify and apple:", quad_corr_in_spotify_apple)
print("Cubic Correlation between spotify and apple:", cubic_corr_in_spotify_apple)
print("Quadratic Correlation between apple and spotify:", quad_corr_in_apple_spotify)
print("Cubic Correlation between apple and spotify:", cubic_corr_in_apple_spotify)
print("Quadratic Correlation between spotify and deez:", quad_corr_in_spotify_deez)
print("Cubic Correlation between spotify and deez:", cubic_corr_in_spotify_deez)
print("Quadratic Correlation between deez and spotify:", quad_corr_in_deez_spotify)
print("Cubic Correlation between deez and spotify:", cubic_corr_in_deez_spotify)
```
